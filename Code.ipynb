## Vintage Analysis

```python
#0: 1-29 days past due 
#1: 30-59 days past due
#2: 60-89 days overdue
#3: 90-119 days overdue
#4: 120-149 days overdue
#5: Overdue or bad debts, write-offs for more than 150 days 
#C: paid off that month 
#X: No loan for the month
cre_rec['vintage'] = cre_rec.groupby('ID')["MONTHS_BALANCE"].transform("min")
cre_rec.head()
```

```python
delinquent_status = ["0", "#1", "#2", "#3", "#4", "#5"]
def calculate_delinquency(group):
    total_loans = len(group)
    delinquent_loans = group["STATUS"].isin(delinquent_status).sum()
    return delinquent_loans / total_loans

vintage_analysis = cre_rec.groupby("vintage", group_keys=False).apply(calculate_delinquency, include_groups=False)
print(vintage_analysis)
```

```python
cre_rec["NEXT_STATUS"] = cre_rec.groupby("ID")["STATUS"].shift(-1)
roll_rate = cre_rec.groupby(["STATUS", "NEXT_STATUS"]).size().reset_index(name = 'count')
print(roll_rate)
```

```python
vintage_analysis.plot(kind="line", marker="o", figsize=(10, 6))
plt.title("Delinquency Rates by Vintage")
plt.xlabel("Vintage (Months from Origination)")
plt.ylabel("Delinquency Rate")
plt.grid()
plt.show()
```

```python
cre_rec.head()
```

```python
cre_rec['label'] = cre_rec.STATUS.apply(lambda x: "bad" if x in delinquent_status else "good")
cre_rec.head()    
```

## Label Encoding

```python
def label(group):
    if(group == 'M'):
        return 1
    return 0    
def labelc(group):
    if(group == 'Y'):
        return 1
    return 0      
df["CODE_GENDER"] = df.CODE_GENDER.apply(lambda x: label(x))  
df["FLAG_OWN_CAR"] = df.FLAG_OWN_CAR.apply(lambda x: labelc(x)) 
df["FLAG_OWN_REALTY"] = df.FLAG_OWN_REALTY.apply(lambda x: labelc(x)) 
df.head()
```

```python
df["final_label"] = df["final_label"].str.strip().str.lower()
def labelt(group):
    if group == 'good':
        return 1
    return 0
df["final_label"] = df["final_label"].apply(labelt)
df.head()
```

## Data Imbalance Techinque: SMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=8)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
rf_clf = RandomForestClassifier()
param_grid_rf = {
    'n_estimators': np.arange(10, 201, 10),  
    'max_features': ['auto', 'sqrt', 'log2'],  
    'max_depth': [None, 10, 20, 30], 
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4],  
    'bootstrap': [True, False]  
}
random_search_rf = RandomizedSearchCV(rf_clf, param_grid_rf, n_iter=50, cv=5, random_state=42, scoring='accuracy')
random_search_rf.fit(x_train_resampled, y_train_resampled)
print("Best Parameters for RandomForest:", random_search_rf.best_params_)
print("Best Accuracy for RandomForest:", random_search_rf.best_score_)
y_pred_rf = random_search_rf.predict(x_test)
accuracy_rf = random_search_rf.score(x_test, y_test)  
print("RandomForest Test Accuracy:", accuracy_rf)
```
